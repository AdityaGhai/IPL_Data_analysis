# Scrapping Players details of IPL 2023

## Importing Libraries

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from tqdm import tqdm
```

## Getting data From the website

```python
def page_content(URL):
    try:
        page_content = requests.get(URL).content
        soup = bs(page_content,"html.parser")
        return soup
    except Exception as e:
        print("Error occured: ",e)
```

```python
url = 'https://www.espncricinfo.com/series/indian-premier-league-2023-1345038/squads'
home = page_content(url)
```

## Getting links of each teams

```python
prefix = 'https://www.espncricinfo.com/'
player_url = []
```

```python
teams = home.find_all('div', class_='ds-flex ds-flex-row ds-space-x-2')
for team in teams:
    link = team.find('a', class_= 'ds-inline-flex ds-items-start ds-leading-none')
    player_url.append(prefix+link.get('href'))
```

## Player Details Scrapping

```python
def image_retriever(player_name):
    player_search_name = "+".join(player_name.split())
    URL = f"https://www.google.com/search?q={player_search_name}+%3A\
    +cricbuzz+profile&tbm=isch&ved=2ahUKEwibjOy215H8AhX0jNgFHfHLCpwQ2\
    -cCegQIABAA&oq=Ben+Stokes+%3A+cricbuzz+profile&gs_lcp=CgNpbWcQAzIHCAAQgAQQGFCgDFj\
    YJGDXKGgAcAB4AIAB7gGIAdkPkgEFMC45LjKYAQCgAQGqAQtnd3Mtd2l6LWltZ8ABAQ&sclient=img&\
    ei=qqOmY5uPKfSZ4t4P8Zer4Ak&bih=657&biw=1366&rlz=1C1CHBF_enIN1009IN1013"
    page = page_content(URL)
    image = page.findAll("img")[1].get("src")
    return image
```

```python
def player_data(s):
    page_content = s.find('div', class_='ds-mb-4')
    team = []
    image = []
    name = []
    bat_style = []
    bowl_style = []
    play_role = []

    team_name = page_content.find('span', class_='ds-text-title-l ds-font-bold ds-text-typo').text[:-5]
    players = page_content.find_all('div', class_='ds-border-line odd:ds-border-r ds-border-b')

    for player in players:
        p = player.find_all('p')
        p_name = player.find('span', class_="ds-text-title-m ds-font-bold ds-text-typo ds-underline ds-decoration-ui-stroke hover:ds-text-typo-primary hover:ds-decoration-ui-stroke-primary ds-block").text    
        team.append(team_name)
        name.append(p_name)
        play_role.append(p[0].text)
        print(p)
        
#         img = image_retriever(p_name)
        image.append("img")                            
        try:
            bat_style.append(p[2].text[9:])
        except Exception as e:
            bat_syle.append('')
        try:
            bowl_style.append(p[3].text[9:])
        except Exception as e:
            bowl_style.append('')
    
    df = pd.DataFrame({
        "Name":name,
        "Image":image,
        "Team":team,
        "Batting_Style":bat_style,
        "Bowling_Style":bowl_style,
        "Playing_Role":play_role
    })
    return df

r = requests.get("https://www.espncricinfo.com//series/indian-premier-league-2023-1345038/chennai-super-kings-squad-1345088/series-squads")

soup2 = bs(r.text, 'html.parser')
df = player_data(soup2)
df
```

```python
player_details = pd.DataFrame()

print(f"{'>'*10} Scrapping players details {'<'*10}")

for link in tqdm(player_url):
    
    try:
        r = requests.get(link)

        soup2 = bs(r.text, 'html.parser')
        df = player_data(soup2)
        player_details = pd.concat([player_details,df],axis=0,ignore_index=True)
    
    except Exception as e:
        continue
    
print("Data Scarpped successfully!!!")
    
```
