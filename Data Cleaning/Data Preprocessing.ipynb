```python
import pandas as pd
```

## Preprocessing Matches data

```python
df_match = pd.read_csv("../Data Collection/match_table.csv")
df_match.head()
```

```python
df_match.shape
```

```python
df_match['Match_Id'] = [f"T20 #{x}" for x in range(1,75)]
```

```python
df_match.drop(columns=['Scorecard'], inplace=True)
df_match.head()
```

```python
df_match = df_match.replace("Titans", "Gujarat Titans")
df_match = df_match.replace("Super Giants", "Lucknow Super Giants")
df_match = df_match.replace("Sunrisers", "Sunrisers Hyderabad")
df_match = df_match.replace("RCB", "Royal Challengers Bangalore")
df_match = df_match.replace("Super Kings", "Chennai Super Kings")
df_match = df_match.replace("Capitals", "Delhi Capitals")
df_match = df_match.replace("Royals", "Rajasthan Royals")
df_match = df_match.replace("KKR", "Kolkata Knight Riders")
df_match = df_match.replace("Mumbai", "Mumbai Indians")

df_match.head()
```

```python
df_match.to_csv('../Data/matches_summary_fact.csv', index=False)
```

```python
match_id_dict = {}

for i, row in df_match.iterrows():
    key1 = row['Team 1'] + " vs " + row['Team 2']
    key2 = row['Team 2'] + " vs " + row['Team 1']
    
    match_id_dict[key1] = row['Match_Id']
    match_id_dict[key2] = row['Match_Id']
    
    
```

## Preprocessing Batting data

```python
df_bat = pd.read_csv("../Data Collection/ipl_bat_summary.csv")
df_bat.head()
```

```python
df_bat["Out/Not_Out"] = df_bat.Dismissal.apply(lambda x: "out" if x != "not out" else "not out")
df_bat.head()
```

```python
df_bat.drop(columns=['Dismissal'], inplace=True)
df_bat.head()
```

```python
df_bat["Batsman"] = df_bat["Batsman"].apply(lambda x: x.replace('â€', ''))
df_bat["Batsman"] = df_bat["Batsman"].apply(lambda x: x.replace('Â', ''))
df_bat["Batsman"] = df_bat["Batsman"].apply(lambda x: x.replace('†', ''))
df_bat["Batsman"] = df_bat["Batsman"].apply(lambda x: x.replace('(c)', ''))
df_bat.head()
```

```python
match_id_dict['Chennai Super Kings vs Gujarat Titans']
```

```python
df_bat["Match_Id"] = df_bat["Match"].map(match_id_dict)
df_bat.head()
```

```python
df_bat.to_csv("../Data/batting_fact_summary.csv", index=False)
```

## Preprocessing Bowling Data

```python
df_bowl = pd.read_csv("../Data Collection/ipl_bowl_summary.csv")
df_bowl.head()
```
